```python
import os
import numpy as np
import math
import torch
from torchvision.io import read_video
from tqdm import tqdm
```

## video.mp4 to 1 second npz files

```python

def process_and_split_videos(input_folder, output_base_folder, existing_files=None, fps=30):
    """
    mp4 비디오 파일을 로드하여 텐서로 변환한 후, 1초씩 나누어 npz 파일로 저장하는 함수.

    Args:
        input_folder (str): mp4 파일이 저장된 입력 폴더 경로.
        output_base_folder (str): 1초짜리 npz 파일을 저장할 기본 출력 폴더 경로.
        existing_files (set): 이미 존재하는 파일 이름 목록.
        fps (int): 초당 프레임 수 (기본값은 30).
    """
    if existing_files is None:
        existing_files = set()

    # Ensure the base output folder exists
    os.makedirs(output_base_folder, exist_ok=True)

    # Iterate over all mp4 files in the input folder with tqdm progress bar
    for filename in tqdm(sorted(os.listdir(input_folder)), desc="Processing videos", unit="file"):
        if filename.endswith('.mp4'):  # mp4 확장자를 가진 파일만 처리
            video_path = os.path.join(input_folder, filename)
            print(video_path)
            # Create the output file prefix (without extension)
            file_prefix = os.path.splitext(filename)[0]

            # Check if the file already exists in the output folder
            if file_prefix in existing_files:
                print(f"File {file_prefix} already exists, skipping...")
                continue

            # Load the video and convert to tensor
            a = read_video(video_path, pts_unit="sec")
            video_tensor = a[0]  # video_tensor의 크기는 (T, C, H, W) 형식

            # Calculate total frames and duration in seconds
            total_frames = video_tensor.shape[0]
            duration = math.ceil(total_frames / fps)
            print(f"Total frames: {total_frames}, Duration: {duration} seconds")

            # 폴더 결정: 각 비디오 파일에 대해 별도의 폴더 생성
            video_output_folder = os.path.join(output_base_folder, file_prefix)
            os.makedirs(video_output_folder, exist_ok=True)

            # Split the video into 1-second clips and save each as .npz
            for i in tqdm(range(duration), desc=f"Splitting {filename}", leave=False):
                start_frame = i * fps
                end_frame = (i + 1) * fps

                # 1초짜리 비디오 클립 추출
                clip = video_tensor[start_frame:end_frame]

                # 파일명을 지정하여 클립을 저장
                output_filename = f"{file_prefix}_{i + 0:02d}.npz"  # 파일명에 인덱스 추가
                output_filepath = os.path.join(video_output_folder, output_filename)

                # npz 파일로 저장
                np.savez_compressed(output_filepath, features=clip.numpy())

            print(f"총 {duration}개의 1초짜리 클립이 {video_output_folder}에 저장되었습니다.")
```

```python
input_folder = "/content/drive/MyDrive/Colab Notebooks/project/AIKU-summer/data/inference/mlb_clips/"
output_folder = "/content/drive/MyDrive/Colab Notebooks/project/AIKU-summer/data/inference/mlb_clips_npz"
process_and_split_videos(input_folder, output_folder)
```

## inference

```python
import torchvision.models as models
import torch.nn as nn

class BinaryActionRecognitionModel(nn.Module):
    def __init__(self):
        super(BinaryActionRecognitionModel, self).__init__()

        # EfficientNet B0 모델을 가져오고, weights를 사용하여 미리 학습된 가중치 로드
        self.efficientnet = models.efficientnet_b0(
            weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1
        )
        self.efficientnet = nn.Sequential(
            *list(self.efficientnet.children())[:-1]
        )  # Feature extractor로 사용

        # Classification Head
        self.fc1 = nn.Linear(1280, 64)  # EfficientNet B0의 출력 크기에 맞게 설정
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(64, 2)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        batch_size, time_steps, c, h, w = x.size()
        # Rescale input
        x = x * 255.0  # Assuming input is in range [0, 1]
        # Reshape input for EfficientNet
        x = x.view(batch_size * time_steps, c, h, w)
        x = self.efficientnet(x)
        x = x.view(batch_size, time_steps, -1)

        # Classification head
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)

        # Global Average Pooling (along time dimension)
        x = x.mean(dim=1)  # GlobalAveragePooling3D의 역할을 하는 부분
        x = x[:, 1]
        return x
```

```python
import torch
import numpy as np
from torchvision import transforms
import os

# 폴더 경로 설정
folder_name = "/content/drive/MyDrive/Colab Notebooks/project/AIKU-summer/data/inference/mlb_clips_npz"

# Resize transform 정의
resize_transform = transforms.Compose([
    transforms.Lambda(lambda x: x.astype(np.uint8)),  # PIL로 변환할 수 있도록 uint8로 변환
    transforms.ToPILImage(),  # numpy 배열을 PIL 이미지로 변환
    transforms.Resize((224, 224)),  # (H, W) 크기로 변환
    transforms.ToTensor()  # 다시 Tensor로 변환하여 (C, H, W) 형식으로 만듬
])

# 모델 초기화 및 체크포인트 불러오기
checkpoint_model = "/content/drive/MyDrive/Colab Notebooks/project/AIKU-summer/checkpoint/epoch0_step16100.pth"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 모델 초기화 (모델 정의는 이미 있다고 가정)
model = BinaryActionRecognitionModel()

# 체크포인트에서 모델 가중치 불러오기
checkpoint = torch.load(checkpoint_model, map_location=device)
model.load_state_dict(checkpoint['model_state_dict'])

# 모델을 평가 모드로 설정
model.eval()

# 모델을 GPU로 이동 (가능한 경우)
model.to(device)
print(device)

# 결과를 저장할 딕셔너리
output_dict = {}

# 폴더 내 모든 파일에 대해 처리
for root, dirs, files in os.walk(folder_name):
    for file_name in tqdm(files, desc=f"Processing files in {root}", unit="file"):
      if file_name.endswith(".npz"):
          file_path = os.path.join(root, file_name)

          # 파일에서 비디오 데이터 로드
          try:
              video_data = np.load(file_path)['features']
          except (EOFError, KeyError) as e:
              print(f"Error loading file {file_path}: {e}")
              continue  # 파일이 손상되었거나 문제 있으면 건너뜀

          # 비디오 데이터가 None이 아닌 경우에만 처리
          if video_data is not None:
              # Transform 적용
              video_data = [resize_transform(frame) for frame in video_data]
              video_data = torch.stack(video_data)

              # 고정된 길이로 맞추기 (padding)
              fixed_length = 30
              fixed_channels = 3

              if video_data.size(0) < fixed_length:
                  padding_frame = torch.zeros((fixed_channels, 224, 224))
                  num_padding_frames = fixed_length - video_data.size(0)
                  padding_frames = [padding_frame] * num_padding_frames
                  padding_frames = torch.stack(padding_frames)
                  video_data = torch.cat((padding_frames, video_data), dim=0)

              # 배치 차원을 추가하여 모델 입력 형식에 맞게 조정
              video_data = video_data.unsqueeze(0)  # (1, fixed_length, 3, 224, 224)

              # GPU로 데이터 이동
              video_data = video_data.to(device)

              # 모델에 입력하여 추론 수행
              with torch.no_grad():
                  output = model(video_data)

              # 딕셔너리에 저장 (key: 파일 이름, value: 모델 출력)
              output_dict[file_name] = output.cpu().numpy()
              print(file_name, output.cpu().numpy())

# 결과 출력
print("Inference completed for all files.")
```

```python
# output_dict를 value 값에 따라 내림차순으로 정렬하고 상위 50개만 선택
sorted_output_dict = dict(sorted(output_dict.items(), key=lambda item: item[1], reverse=True)[:100])

# 상위 50개 항목 출력
for file_name, value in sorted_output_dict.items():
    print(f"{file_name}: {value}")
```

```python
from moviepy.video.io.VideoFileClip import VideoFileClip
from tqdm import tqdm
import os

# 비디오 클립이 저장된 폴더 경로
base_folder = "/content/drive/MyDrive/Colab Notebooks/project/AIKU-summer/data/inference/mlb_clips/"
# 클립을 저장할 폴더 경로
output_folder = "/content/drive/MyDrive/Colab Notebooks/project/AIKU-summer/data/inference/mlb_clips_combine/"

# 출력 폴더가 존재하지 않으면 생성
os.makedirs(output_folder, exist_ok=True)

for key in tqdm(filtered_dict.keys(), desc="Processing clips", unit="clip"):
    # 파일 이름에서 'mlb_clip_00_31.npz'처럼 00과 31 부분 추출
    clip_prefix = key.split('_')[2]  # '00_31' 중 '00' 추출
    start_time = int(key.split('_')[3].split('.')[0])  # '00_31' 중 '31' 추출
    end_time = start_time + 1  # 종료 시간은 시작 시간에 1초 더한 값

    # 비디오 파일 경로 설정
    video_file = os.path.join(base_folder, f"mlb_clip_{clip_prefix}.mp4")

    # 비디오 파일이 존재하는지 확인
    if os.path.exists(video_file):
        # 비디오 파일 로드
        print(video_file)
        video = VideoFileClip(video_file)

        # 비디오에서 필요한 구간 추출
        clip = video.subclip(start_time, end_time)

        # 출력 파일 경로 설정
        output_file = os.path.join(output_folder, f"{key}.mp4")

        # 클립을 개별 파일로 저장
        clip.write_videofile(output_file, codec="libx264", audio_codec="aac")

        # 리소스 해제
        video.close()
    else:
        print(f"Video file {video_file} not found. Skipping this clip.")
```

```python
from moviepy.editor import VideoFileClip, concatenate_videoclips
import os

# 비디오 클립이 저장된 폴더 경로
input_folder = "/content/drive/MyDrive/Colab Notebooks/project/AIKU-summer/data/inference/mlb_clips_combine/"
output_file = "/content/drive/MyDrive/Colab Notebooks/project/AIKU-summer/data/inference/mlb_combined_output.mp4"

# 폴더 내 모든 MP4 파일을 불러와서 하나의 리스트에 저장
video_clips = []
for filename in os.listdir(input_folder):
    if filename.endswith(".mp4"):
        file_path = os.path.join(input_folder, filename)
        video_clips.append(VideoFileClip(file_path))

# 모든 비디오 클립을 하나로 결합
final_clip = concatenate_videoclips(video_clips)

# 결합된 비디오를 파일로 저장
final_clip.write_videofile(output_file, codec="libx264", audio_codec="aac")

# 리소스 해제
for clip in video_clips:
    clip.close()
```
