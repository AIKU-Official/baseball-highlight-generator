```python
import torchvision
import av
print(torchvision.__version__)
print(av.__version__)
torchvision.set_video_backend('pyav')
from torchvision.io import read_video

def video_to_tensor(video_file):
    # Extract audio from video using FFmpeg
    # Output audio in wav format and pipe the output to stdout

    a = read_video(video_file, pts_unit="sec")
    video_tensor = a[0]
    return video_tensor
```

```python
import numpy as np
import os
import torch
from tqdm import tqdm
def process_videos(input_folder, output_folder):
    # Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    # Iterate over all files in the input folder with tqdm progress bar
    for filename in tqdm(sorted(os.listdir(input_folder)), desc="Processing videos", unit="file"):
        if filename.endswith(('.mp4')):  # Adjust based on your video formats
            video_path = os.path.join(input_folder, filename)

            # Create the output filename with .npz extension
            output_filename = os.path.splitext(filename)[0] + '.npz'
            output_path = os.path.join(output_folder, output_filename)

            # Check if the file already exists, if so, skip processing
            if os.path.exists(output_path):
                print(f"File {output_filename} already exists, skipping...")
                continue

            # Process the video and save to .npz
            video_tensor = video_to_tensor(video_path)

            # Save the numpy array (tensor) to a .npz file
            np.savez_compressed(output_path, features=video_tensor.numpy())

            print(f"Processed and saved: {output_path}")
```

```python
# Example usage:
input_folder = "/content/drive/MyDrive/Colab Notebooks/project/AIKU-summer/mlb-dataset/continuous"
output_folder = "/content/drive/MyDrive/Colab Notebooks/project/AIKU-summer/data/video_features"
```

```python
process_videos(input_folder, output_folder)
```
